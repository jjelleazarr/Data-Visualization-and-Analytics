# Exploratory Data Analysis and Visualization


## Reading in the data
First we import the Python packages we are going to use. Then we use Pandas to load in the dataset as a data frame.


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
df1 = pd.read_csv('/Users/jjelleazarr/Documents/GitHub/Data-Visualization-and-Analytics/Lectures (Prelims)/pokemon.csv', index_col=0, encoding='ISO-8859-1')
```

## Examine the dataset

```python
print(df1.head())
print(df1.describe())
```

## Plotting a histogram in Python


```python
g = plt.hist(df1['Speed'], histtype='bar', ec='black')
g = plt.xlabel('Pokemon Speed') 
g = plt.ylabel('Frequency')
plt.show()
```

## Bins
There are a couple of ways to manipulate bins in matplotlib. Here, I specified where the edges of the bars of the histogram are; the bin edges.

```python
bin_edges = [0,20,40,60,80,100,120,140,160,180] # specifying the edges of the bins
g = plt.hist(df1['Speed'], bins=bin_edges, histtype='bar', ec='black')
g = plt.xlabel('Pokemon Speed') 
g = plt.ylabel('Frequency')
plt.show()
```

You could also specify the number of bins, and Matplotlib will automatically generate a number of evenly spaced bins.


```python
g = plt.hist(df1['Speed'], bins=20, histtype='bar', ec='black') # specifying the number of bins
g = plt.xlabel('Pokemon Speed') 
g = plt.ylabel('Frequency')
plt.show()
```

# Plotting with Seaborn
Histograms allow you to plot the distributions of numeric variables.


```python
sns.set_style("whitegrid") # setting the style
sns.displot(df1.Speed)
```

# Other types of graphs: Creating a scatter plot


```python
sns.lmplot(x='Attack', y='Defense', data=df1)
```

Seaborn doesn't have a dedicated scatter plot function. We used Seaborn's function for fitting and plotting a regression line; hence lmplot()  However, Seaborn makes it easy to alter plots. To remove the regression line, we use the fit_reg=False command


```python
sns.lmplot(x='Attack', y='Defense', data=df1, fit_reg=False)
```

# The hue function
Another useful function in Seaborn is the hue function, which enables us to use a variable to colour code our data points.

```python
sns.lmplot(x='Attack', y='Defense', data=df1, fit_reg=False, hue='Generation')
```

## Factor plots (Catplot)
Make it easy to separate plots by categorical classes.

```python
g = sns.catplot(x='Type 1', y='Attack', data=df1, hue='Legendary', col='Legendary', kind='swarm')
g.set_xticklabels(rotation=-45)
```

# A box plot

```python
sns.boxplot(df1)
```

The total, generation, and legendary entries are not combat stats so we should remove them. Pandas makes this easy to do, we just create a new dataframe We just use Pandas .drop() function to create a dataframe that doesnt include the variables we dont want.


```python
stats_df = df1.drop(['Total', 'Generation', 'Legendary'], axis=1)
sns.boxplot(data=stats_df)
```

## Seaborns theme
Seaborn has a number of themes you can use to alter the appearance of plots. For example, we can use darkgrid to add grid lines to our boxplot.


```python
stats_df = df1.drop(['Total', 'Generation', 'Legendary'], axis=1)
sns.set_style("darkgrid")
sns.boxplot(data=stats_df)
```

# Violin plots
Violin plots are useful alternatives to box plots. They show the distribution of a variable through the thickness of the violin. Here, we visualise the distribution of attack by Pokmon's primary type:


```python
sns.violinplot(x='Type 1', y='Attack', data=df1)
```

## Seaborns colour palettes
Seaborn allows us to easily set custom colour palettes by providing it with an ordered list of colour hex values. We first create our colours list.


```python
type_colors = ['#78C850', '#F08030', '#6890F0', '#A8B820', '#A040A0', '#F8D030', '#E0C068', '#EE99AC', '#C03028', '#F85888', '#B8A038', '#705898', '#98D8D8', '#7038F8', '#705848', '#A890F0', '#B8B8D0', '#A6A878']
```

Then we just use the palette= function and feed in our colours list.


```python
sns.violinplot(x='Type 1', y='Attack', data=df1, palette=type_colors, hue='Type 1')
```

## Swarm Plot
Because of the limited number of observations, we could also use a swarm plot. Here, each data point is an observation, but data points are grouped together by the variable listed on the x-axis.


```python
sns.swarmplot(x='Type 1', y='Attack', data=df1, palette=type_colors, hue='Type 1')
```

## Overlapping Plots
Both of these show similar information, so it might be useful to overlap them.

```python
plt.figure(figsize=(14,6))
sns.violinplot(x='Type 1', y='Attack', data=df1, palette=type_colors, hue='Type 1', inner=None)
sns.swarmplot(x='Type 1', y='Attack', data=df1, color='k', alpha=0.6)
plt.title('Attack by Primary Type')
```

## Data wrangling with Pandas
What if we wanted to create such a plot that included all of the other stats as well? In our current dataframe, all of the variables are in different columns. If we want to visualise all stats, then well have to melt the dataframe.

```python
stats_df = df1.drop(['Total', 'Generation', 'Legendary'], axis=1)
melted_df = pd.melt(stats_df, id_vars=['Name', 'Type 1', 'Type 2'], var_name='Stat')
print(melted_df.head())
```

```python
sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1', dodge=True)
```

This graph could be made to look nicer with a few tweaks.


```python
plt.figure(figsize=(14,6))
sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1', palette=type_colors, dodge=True)
plt.ylim(0, 260)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

## Plotting all data: Empirical cumulative distribution functions (ECDFs)
An alternative way of visualising a distribution of a variable in a large dataset is to use an ECDF.

```python
x = np.sort(df1['Attack'])
y = np.arange(1, len(x)+1) / len(x)
g = plt.plot(x, y, marker='.', linestyle='none')
g = plt.xlabel('Attack')
g = plt.ylabel('ECDF')
plt.show()
```

You can also plot multiple ECDFs on the same plot.

```python
x1 = np.sort(df1['Attack'])
x2 = np.sort(df1['Defense'])
x3 = np.sort(df1['Speed'])
y = np.arange(1, len(x1)+1) / len(x1)
plt.plot(x1, y, marker='.', linestyle='none', label='Attack')
plt.plot(x2, y, marker='.', linestyle='none', label='Defense')  
plt.plot(x3, y, marker='.', linestyle='none', label='Speed')
plt.xlabel('Stats')
plt.ylabel('ECDF')
plt.legend()
plt.show()
```

## Heatmaps
Useful for visualising matrix-like data.


```python
# Select only numeric columns for correlation
numeric_df = stats_df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
```
